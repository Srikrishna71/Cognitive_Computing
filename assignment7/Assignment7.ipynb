```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
r_no = 102317291
np.random.seed(r_no)
sales_data = np.random.randint(1000,5001,size = (12,4))
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
categories = ["Electronics","Clothing","Home & Kitchen","Sports"]
df = pd.DataFrame(sales_data,columns = categories,index = months)
df
```

```python
# Display first 5 rows
print(df.head())
print()
# Summary statistics
print(df.describe())
print()
total_sales_category = df.sum()
print(total_sales_category)
print()
df["Total Sales"] = df.sum(axis=1)
df
print()
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
df
print()
if r_no % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.90
else:
    df["Clothing"] = df["Clothing"] * 0.85
df
print()
```

```python
# Line plot for sales trends
plt.figure(figsize = (10,6))
for category in categories:
    plt.plot(df.index,df[category],marker = "o",label = category)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.show()
# Box plot for sales distribution
plt.figure(figsize = (8,6))
sns.boxplot(data = df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
array = np.array([[1,-2,3],[-4,5,-6]])
# i. Element-wise absolute value
abs_values = np.abs(array)
print("Absolute Values:\n",abs_values)
# ii. Percentiles
percentiles = np.percentile(array,[25,50,75],axis = None)  # Flattened
percentiles_columns = np.percentile(array,[25,50,75],axis = 0)  # Per column
percentiles_rows = np.percentile(array,[25,50,75],axis = 1)  # Per row
print("Percentiles (Flattened) :",percentiles)
print("Percentiles (Per Column) :",percentiles_columns)
print("Percentiles (Per Row) :",percentiles_rows)
# iii. Mean, Median, Standard Deviation
mean_all = np.mean(array)
median_all = np.median(array)
std_all = np.std(array)
mean_column = np.mean(array,axis = 0)
median_column = np.median(array,axis = 0)
std_column = np.std(array,axis = 0)
mean_row = np.mean(array,axis = 1)
median_row = np.median(array,axis = 1)
std_row = np.std(array,axis = 1)
print("Mean (All) :",mean_all)
print("Median (All) :",median_all)
print("Standard Deviation (All) :",std_all)
print("Mean (Column) :",mean_column)
print("Median (Column) :",median_column)
print("Standard Deviation (Column) :",std_column)
print("Mean (Row) :",mean_row)
print("Median (Row) :",median_row)
print("Standard Deviation (Row) :",std_row)
```

```python
a = np.array([-1.8,-1.6,-0.5,0.5,1.6,1.8,3.0])
f = np.floor(a)
c = np.ceil(a)
t = np.trunc(a)
r = np.round(a)
print("Floor :", f)
print("Ceiling :", c)
print("Truncated :", t)
print("Rounded :", r)
```

```python
def swap_list_elements(l,a,b):
    temp = l[a]
    l[a] = l[b]
    l[b] = temp
    return l
l = [1,2,3,4,5]
print("Before Swap :",l)
print("After Swap :",swap_list_elements(l,2,4))
```

```python
s = {10, 20, 30, 40, 50}
print("Original set:", s)
# Convert set to list
l = list(s)
i, j = 1, 3
l[i], l[j] = l[j], l[i]
# Convert back to set
swaps = set(l)
print("Set after swapping 1 and 3 elements:", swaps)
```
